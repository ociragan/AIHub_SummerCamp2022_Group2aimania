```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
```

```python
cat_path = "/content/drive/MyDrive/PetImages/Cat"
dog_path = "/content/drive/MyDrive/PetImages/Dog"
```

```python
data = []
imagesize = 160

for i in range(0,2510):
  try:
    pict_nr = str(i) + ".jpg"
    path = os.path.join (cat_path, pict_nr)
    img_array = cv2.imread(path)
    resized_array = (cv2.resize(img_array, dsize=[imagesize, imagesize]))/255
    data.append([0, resized_array])
    if i % 200 == 0:
      print(f" cat: {i}")
  except Exception as ex:
    print(f"error reading file: {i}, {ex}")
    pass 
    
for i in range(0,2510):
  try:
    pict_nr = str(i) + ".jpg"
    path = os.path.join (dog_path, pict_nr)
    img_array = cv2.imread(path)
    resized_array = (cv2.resize(img_array, dsize=[imagesize, imagesize]))/255
    data.append([1, resized_array])
    if i % 200 == 0:
      print(f" dog: {i}")
  except Exception as ex:
    print(f"error reading file: {i}, {ex}")
    pass 
print(len(data))
```

```python
import random
random.shuffle(data)
for sample in data[:20]:
  if sample[0] == 0:
    print("cat")
  else:
    print("dog")
```
